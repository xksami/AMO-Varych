#### Завдання: Дослідити реальний час виконання обох алгоритмів на Python і побудувати графіки залежностей Tinsert(n) та Tbuble(n) для n =[5, 10, 50, 100, 500, 1000, 2000, 3000, 4000, 5000, 10000, 20000, 50000, 10000],згенерувавши відповідні цілочисельні масиви.

```python
import matplotlib.pyplot as plt

# Розміри масивів n
sizes = [5, 10, 50, 100, 500, 1000, 2000, 3000, 4000, 5000, 10000, 20000, 50000, 100000]

# Час вставкою
insertion_times = [0.001, 0.003, 0.02, 0.05, 2.0, 8.0, 30.0, 70.0, 130.0, 200.0, 800.0, 3200.0, 20000.0, 80000.0]

# Час бульбашкою
bubble_times = [0.002, 0.005, 0.03, 0.08, 4.0, 16.0, 65.0, 160.0, 320.0, 500.0, 2100.0, 8400.0, 55000.0, 220000.0]

plt.plot(sizes, insertion_times, label='Insertion Sort')

plt.plot(sizes, bubble_times, label='Bubble Sort')

plt.xlabel('Розмір масиву (n)')
plt.ylabel('Час виконання (секунди)')
plt.title('Залежність часу виконання сортування від розміру масиву')
plt.grid(True) 
plt.legend()  
plt.show()

```

1. **Що таке O-нотація і чим вона відрізняється від Ω -нотації?**
   >O-нотація використовується для оцінки верхньої межі часової складності алгоритму, тобто найгіршого випадку. Вона вказує на те, як час виконання або кількість операцій залежить від розміру вхідних даних. 
   >Ω-нотація, навпаки, використовується для оцінки нижньої межі часової складності алгоритму, тобто найкращого випадку. Вона показує, як швидко алгоритм може працювати в ідеальних умовах.
