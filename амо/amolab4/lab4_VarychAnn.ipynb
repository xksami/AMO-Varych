#### Завдання 1: Обчислення факторіалу.

#### Постановка задачі: 
>Реалізувати процедуру обчислення факторіалу за допомогою цикла for. Оцінити асимптотичну складність алгоритму.


```python
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

n = 7
print("Факторіал числа", n, "дорівнює", factorial(n))
```

#### Завдання 2: Рекурсивний алгоритм.

#### Постановка задачі:
>Оцінити асимптотичну складність рекурсивного алгоритму обчислення факторіалу.

```python
def factorial_recurs(n):
    if n == 0:
        return 1
    else:
        return n * factorial_recurs(n - 1)

n = 9
print("Факторіал числа", n, "дорівнює", factorial_recurs(n))
```

#### Завдання 3: Обчислення чисел Фібоначчі.

#### Постановка задачі:
> Оцінити асимптотичну складність рекурсивного алгоритму обчислення n-го числа Фібоначчі.

```python
def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)

n = 16
print("n-е число фібоначчі:", fibonacci(n))
```

#### Завдання 4: Сортування злиттям.

#### Постановка задачі:
>Оцінити асимптотичну складність алгоритму сортування злиттям, використовуючи основну теорему рекурсії.

```python
def merge(left, right):
   
    result = []
    i, j = 0, 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result += left[i:]
    result += right[j:]
    return result

def merge_sort(arr):

    if len(arr) <= 1:
        return arr
    middle = len(arr) // 2
    left = merge_sort(arr[:middle])
    right = merge_sort(arr[middle:])
    return merge(left, right)


new_arr = [9, 2, 7, 4, 1, 8, 3, 6, 5]

sorted_new_arr = merge_sort(new_arr)
print("відсортований масив:", sorted_new_arr)
```

>Щоб записати асимптотичну складність в нотації $O()$, ми шукаємо верхню межу функції $F(N)$. Для цього знаходимо найвищий степінь члену у функції, оскільки це визначає основний вплив на складність при великих значеннях $N$. функція має вигляд $F(N)=N^3+7N^2-14N$. Найвищий степінь у цій функції - $N^3$. Тому асимптотична складність буде виглядати: F(N)=O(N^3)

>У моєму випадку функція має вигляд $F(N)=1.01^N+N^{10}$. При досить великих значеннях $N$ перший член $1.01^N$ зростає експоненціально, тоді як другий член $N^{10}$ зростає значно повільніше.Тому основним впливом на складність буде член $1.01^N$. Таким чином, асимптотична складність може бути F(N)=O(1.01^N)
